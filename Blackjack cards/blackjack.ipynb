```python
import random

# Function to deal random card
def deal_cards():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    card = random.choice(cards)
    return card

# Function to calculate score
def calculate(cards):
    # Blackjack (Ace + 10 or picture card)
    if sum(cards) == 21 and len(cards) == 2:
        return 0
    
    # Adjust Ace (11 â†’ 1) if score is over 21
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    
    return sum(cards)
# Compare results
def compare(user, computer):
    if user == computer:
        return "Draw ðŸ™ƒ"
    elif computer == 0:
        return "Lose, opponent has Blackjack ðŸ˜±"
    elif user == 0:
        return "Win with a Blackjack ðŸ˜Ž"
    elif user > 21:
        return "You went over. You lose ðŸ˜­"
    elif computer > 21:
        return "Opponent went over. You win ðŸ˜"
    elif user > computer:
        return "You win ðŸ˜ƒ"
    else:
        return "You lose ðŸ˜¤"


usercards = []
computercards = []
isgameover = False

# Deal 2 cards to each player
for _ in range(2):
    usercards.append(deal_cards())
    computercards.append(deal_cards())

# Game loop
while not isgameover:
    userscore = calculate(usercards)
    computerscore = calculate(computercards)

    print(f"Your cards: {usercards}, current score: {userscore}")
    print(f"Computer's first card: {computercards[0]}")

    if userscore == 0 or computerscore == 0 or userscore > 21:
        isgameover = True
    else:
        usershoulddeal = input("Type 'y' to get another card, 'n' to pass: ")
        if usershoulddeal == "y":
            usercards.append(deal_cards())
        else:
            isgameover = True

# Computer plays (draws until score >= 17)
while computerscore != 0 and computerscore < 17:
    computercards.append(deal_cards())
    computerscore = calculate(computercards)

print(f"Your final hand: {usercards}, final score: {userscore}")
print(f"Computer's final hand: {computercards}, final score: {computerscore}")


   
print(compare(userscore, computerscore))
```
